```python
df2 = df.groupby('Continent')[df.columns[]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
df2
```

```python
df3 = df2.transpose()
```

```python
plt.figure(figsize=(12, 7))
df3.plot()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#The figure above gives the dates in a an order that is the reverse of what i wanted. To correct this, the slicing changes. An alternative way to approach this is print out the columns and then copy and paste them as they are to avoid the reverse order. 

df2 = df.groupby('Continent')[df.columns[12:4:-1]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
df3 = df2.transpose()
```

```python
plt.figure(figsize=(12, 7))
df3.plot()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
